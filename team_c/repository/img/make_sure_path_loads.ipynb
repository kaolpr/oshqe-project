```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt


df = pd.read_csv('combined_outer_path.csv')
df.head()
```

```python
plt.plot(df['x'], df['y'], 'o', markersize=1, alpha=0.5)
plt.gca().set_aspect('equal', adjustable='box')
```

```python
df_downsampled = df[::50]
plt.plot(df_downsampled['x'], df_downsampled['y'], markersize=1, alpha=0.5)
print("N points: ", len(df_downsampled))
plt.gca().set_aspect('equal', adjustable='box')
```
